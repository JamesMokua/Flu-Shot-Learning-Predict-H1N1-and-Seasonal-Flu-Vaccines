```python
# importng necessary libraries
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 100)
```

```python
# Reading the necessary csv files
features_df = pd.read_csv("training_set_features.csv", index_col="respondent_id")
labels_df = pd.read_csv("training_set_labels.csv", index_col="respondent_id")


# Displaying the data
print("features_df.shape", features_df.shape)
features_df.head()
```

```python
print("labels_df.shape", labels_df.shape)
labels_df
```

```python
# checking whether the rows match
# no output means that the rows match
np.testing.assert_array_equal(features_df.index.values, labels_df.index.values)
```

```python
%matplotlib inline
fig, ax = plt.subplots(2, 1, sharex=True)

n_obs = labels_df.shape[0]

(labels_df['h1n1_vaccine']
    .value_counts()
    .div(n_obs)
    .plot.barh(title="Proportion of H1N1 Vaccine", ax=ax[0])
)
ax[0].set_ylabel("h1n1_vaccine")

(labels_df['seasonal_vaccine']
    .value_counts()
    .div(n_obs)
    .plot.barh(title="Proportion of Seasonal Vaccine", ax=ax[1])
)
ax[1].set_ylabel("seasonal_vaccine")

fig.tight_layout()
```

```python
# combining the labels df and 
# the features df
joined_df = features_df.join(labels_df)
print(joined_df.shape)
joined_df.head()
```

```python
# getting counts of observations for each of the two variables
counts = (joined_df[['h1n1_concern', 'h1n1_vaccine']]
              .groupby(['h1n1_concern', 'h1n1_vaccine'])
              .size()
              .unstack('h1n1_vaccine')
         )
counts
```

```python
# convert values to graph to get a clearer picture
ax = counts.plot.barh()
ax.invert_yaxis()
ax.legend(
    loc='center right', 
    bbox_to_anchor=(1.3, 0.5), 
    title='h1n1_vaccine'
)

# This graph is not so good bc we can't tell what the propotions are
```

```python
# Better graph using the rate of vaccination for each 
# h1n1_concern level
h1n1_concern_counts = counts.sum(axis='columns')
h1n1_concern_counts
```

```python
# Get the propotions
props = counts.div(h1n1_concern_counts, axis='index')
props
```

```python
# plot a graph
ax = props.plot.barh()
ax.invert_yaxis()
ax.legend(
    loc='center left', 
    bbox_to_anchor=(1.05, 0.5),
    title='h1n1_vaccine'
)

#  better graph that shows propotions more clearly
```

```python
# Convert to a stacked chart bc the values add up to 1
# and it's easier to read
ax = props.plot.barh(stacked=True)
ax.invert_yaxis()
ax.legend(
    loc='center left', 
    bbox_to_anchor=(1.05, 0.5),
    title='h1n1_vaccine'
)
```
